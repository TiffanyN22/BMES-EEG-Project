```python
#load libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#basic line graph
#documentation: https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html 
x=[0,1,2,3,4]
y=[0,2,4,6,8]

#resive
plt.figure(figsize=(5,3), dpi=150) #dpi: pixels per inch, often higher

#plt.plot(x,y) 
#plt.plot(x,y, label="2x", color='red', linewidth=2, marker='.', markersize=10, markeredgecolor='blue', linestyle='--')  #can also pass in hex for color, marker for poitns
#shorthand: 'color maker line', check documentation for markers
plt.plot(x,y,'bx--', label='2x')

#line number 2 
x2 = np.arange(0,4.5,0.5) #array from 0-3.5, ticks are 0.5 each time
#plt.plot(x2, x2**2, 'r', label='X^2', marker='.', markersize=10,)
#progression -> plot first 5 (doesn't show 6) one way 
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

#title and axis label text
#plt.title("Our First Graph!")
plt.title("Our First Graph!", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20}) #add dictionary for font
#font list: https://jonathansoma.com/lede/data-studio/matplotlib/list-all-fonts-available-in-matplotlib-plus-samples/
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

#axis labeling
plt.xticks([0,1,2,3,4]) #labels for x axis
plt.yticks([0,2,2.5,4,6,8,10]) #graph will resize based on ticks, puts at right spot even if not evenly spaced out

#add legend
plt.legend() #pass in label to plot function

#save graph
plt.savefig('mygraph.png', dpi=300) #save into same directory

#correct way to show plot without other lines
plt.show() 
```
